```python
import pandas as pd
import simplejson as json

from matplotlib import pyplot as plt
from matplotlib import gridspec
from mpl_toolkits import mplot3d
import seaborn as sns

import scipy.cluster.hierarchy as sch

import numpy as np

#from pymol import cmd
#import pymol
import os, shutil, itertools

%config Completer.use_jedi = False
```

```python
def get_pharmacophore (json_file=None):
    
    color_code={ 'Hydrophobic':        'green',\
                 'HydrogenAcceptor':   'orange',\
                 'HydrogenDonor':      'white',\
                 'Aromatic':           'purple',\
                 'NegativeIon':        'red',\
                 'PositiveIon':        'navy',\
                 'InclusionSphere':    'gray',\
                 'Other':              'yellow',\
                 'PhenylalanineAnalog':'pink',\
                 'LeuValAnalog':       'pink' \
                 }
    with open (json_file, 'r') as file:
        data = json.load(file)
        table=pd.DataFrame(data.get('points'))
    for i in table.index:
        table.loc[i,'color']=color_code[table.loc[i,'name']]
    
    lig=data.get('ligand')
    receptor=data.get('receptor')
    return table, lig, receptor


def plot_pharmacophore(table=None,enabled=None):
    if enabled==True:
        table=table[table['enabled']==True]
    elif enabled==False:
        table=table[table['enabled']==False]
    else:
        table=table
    
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection = '3d')

    for point in table.index:
        ax.scatter(table.loc[point,'x'],table.loc[point,'y'],table.loc[point,'z'],
                     s=table.loc[point,'radius']*100,
                     color=table.loc[point,'color'],edgecolors='k')

        if 'vector_on' in table.columns and table.loc[point,'vector_on']==0:
            origin=[table.loc[point,'x'],table.loc[point,'y'], table.loc[point,'z']]
            vec=[table.loc[point,'svector']['x'],table.loc[point,'svector']['y'],table.loc[point,'svector']['z']]
            ax.quiver(origin[0],origin[1],origin[2], vec[0],vec[1],vec[2], color='k',alpha=0.5,linestyle='-',linewidth=2)

    ax.set_xlabel("X coors",fontsize=20,fontweight='bold')
    ax.set_ylabel("Y coors",fontsize=20,fontweight='bold')
    ax.set_zlabel("Z coors",fontsize=20,fontweight='bold')
    plt.tick_params ('both',width=2,labelsize=12)
    plt.tight_layout()

    return fig



def save_p4_pymol(table,out_name='p4.pse',select='all'):
    
    if select=='cluster':
        for point in table.index:
            cmd.pseudoatom(object=int(table.loc[point,'cluster']),resn=table.loc[point,'name'],
                               resi=point, chain='P', elem='PS',label=int(table.loc[point,'cluster']),
                               vdw=table.loc[point,'radius'], hetatm=1, color=table.loc[point,'color'],b=table.loc[point,'weight'],
                           q=table.loc[point,'balance'],pos=[table.loc[point,'x'],table.loc[point,'y'],table.loc[point,'z']])

            cmd.group(table.loc[point,'name'], '*')
    
    elif select=='concensus':
        for point in table.index:
            cmd.pseudoatom(object=table.loc[point,'name'],resn=table.loc[point,'name'],
                           resi=point, chain='P', elem='PS',label=table.loc[point,'name'],
                           vdw=table.loc[point,'radius'], hetatm=1, color=table.loc[point,'color'],b=table.loc[point,'weight'],
                           q=table.loc[point,'balance'],pos=[table.loc[point,'x'],table.loc[point,'y'],table.loc[point,'z']])
            
            cmd.group('Concensus', '*')
        
    elif select=='all':
        for point in table.index:
            cmd.pseudoatom(object=table.loc[point,'name'],resn=table.loc[point,'name'],
                           resi=point, chain='P', elem='PS',label=table.loc[point,'name'],
                           vdw=table.loc[point,'radius'], hetatm=1, color=table.loc[point,'color'],
                           pos=[table.loc[point,'x'],table.loc[point,'y'],table.loc[point,'z']])
    
    else:
        print("parameter select must be 'all', 'cluster' or 'concensus'")
    cmd.save(out_name,format='pse')
    cmd.remove('all')
    cmd.reinitialize(what='everything')
    
    
def get_distances_matrix(table=None):
    matrix=pd.DataFrame()
    for i in table.index:
        for j in table.index:
            coord_A=np.array((table.loc[i,'x'],table.loc[i,'y'],table.loc[i,'z']))
            coord_B=np.array((table.loc[j,'x'],table.loc[j,'y'],table.loc[j,'z']))
            dist_numpy = np.linalg.norm(coord_A-coord_B)        
            matrix.loc[i,j]=dist_numpy

    return matrix
    

def plot_clusters(table=None):
    
    m=get_distances_matrix(table)
    fig=sns.clustermap (m,method='complete',figsize=(10,10),xticklabels=m.columns, yticklabels=m.index,cmap='RdBu',cbar_kws=dict(label='Distance',shrink=1,orientation='vertical',spacing='uniform',pad=0.02))
    
    return fig
    
def get_clusters(table=None,plot=False):
    
    data=table.copy()
    m=get_distances_matrix(data)
    d = sch.distance.pdist(m)
    L = sch.linkage(d, method='complete')
    clusters = sch.fcluster(L, 0.175*d.max(), 'distance')

    for i,cluster in enumerate(clusters):
        data.loc[m.index[i],'cluster']=cluster
    
    balance=[]
    for i in m.index:
        weight=len([d for d in m[i] if d <= 1.5])
        balance.append(weight)
        data.loc[i,'weight']=weight
        
    if len(data) > 1:
        bal=[(i-min(balance))/(max(balance)-min(balance)) for i in balance]
    else:
        bal=[0]
    
    data['balance']=bal

    
    if plot==True:
        fig=sns.clustermap (m,method='complete',figsize=(10,10),xticklabels=m.columns, yticklabels=m.index,
                            cmap='RdBu',cbar_kws=dict(label='Distance',shrink=1,orientation='vertical',spacing='uniform',pad=0.02))
    
    return data

def get_centroid(table):
    length = len(table.index)
    sum_x = np.sum(table['x'])
    sum_y = np.sum(table['y'])
    sum_z = np.sum(table['z'])
    return sum_x/length, sum_y/length, sum_z/length


def get_CM_radius (table):
    
    CM = np.average([table['x'],table['y'],table['z']], axis=1, weights=table['weight'])
    radius = max([np.linalg.norm(CM - np.array((table.loc[i,'x'],table.loc[i,'y'],table.loc[i,'z']))) for i in table.index])/2
    
    if radius==0:
            radius=0.5
    
    return CM,radius


def generate_json (table=None,out_name='Concensus.json'):
    points=[]
    for i in table.index:
        points.append(Concensus.loc[i].to_dict())
    
    p={"points" :points}
    with open(out_name, 'w') as fp:
        json.dump(p, fp)
```

```python
results=[]
for file in os.listdir('../Structures/Pharmacophore/'):
    if '.json' in file:
        p4,lig,rec=get_pharmacophore('../Structures/Pharmacophore/'+file)
        results.append(p4)
TABLE = pd.concat(results)
TABLE.sort_values('name',inplace=True)
TABLE.reset_index(drop=True,inplace=True)
```

## Clustering of P4

```python
Descriptors=Enabled.groupby('name')
```

```python
Descriptors.groups.keys()
```

```python
Concensus=pd.DataFrame()
weights=[]
balances=[]
index=1
for group in Descriptors.groups:
    if group in ['HydrogenAcceptor', 'HydrogenDonor', 'Hydrophobic']:
        Clusters=get_clusters(Descriptors.get_group(group))
        #plot_clusters(Clusters)
        save_p4_pymol (table=Clusters,out_name='../p4_concenso/'+group+'_clusters.pse',select='cluster')
        Points=Clusters.groupby('cluster')
        for g in Points.groups:
            clus=Points.get_group(g)
            CM,radius=get_CM_radius(clus)
            Concensus.loc[index,'name']=group
            Concensus.loc[index,'cluster']=g
            Concensus.loc[index,'x']=CM[0]
            Concensus.loc[index,'y']=CM[1]
            Concensus.loc[index,'z']=CM[2]
            Concensus.loc[index,'radius']=radius
            Concensus.loc[index,'color']=clus.loc[clus.index[0],'color']
            Concensus.loc[index,'weight']=clus['weight'].mean()
            Concensus.loc[index,'balance']=clus['balance'].mean()
            
            index=index+1
```
